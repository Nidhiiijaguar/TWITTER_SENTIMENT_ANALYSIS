```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as  plt
import datetime
from dateutil.relativedelta import relativedelta
from datetime import date
import math
```

```python
df=pd.read_csv("covid19_tweets.csv")
```

```python
df
```

```python
df.shape
```

```python
df["Total_Tweets"]=df["user_name"].map(df["user_name"].value_counts())
df
```

```python
df["date"]=pd.to_datetime(df["date"])
Tdays=math.ceil((df.iloc[df.shape[0]-1]["date"]-df.iloc[0]["date"]) / np.timedelta64(1,'D'))
Tdays
```

```python
df["Tweets per day"]=df["Total_Tweets"]/Tdays
df
```

```python
Activecondition=[
    (df["Tweets per day"]<=0.1),
    (df["Tweets per day"]>0.1) & (df["Tweets per day"]<=0.7),
    (df["Tweets per day"]>0.7) & (df["Tweets per day"]<=2),
    (df["Tweets per day"]>2) & (df["Tweets per day"]<=5),
    (df["Tweets per day"]>5)
]
Activity=["Poor Active","Less Active","Moderate","Highly Active","Extremely Active"]
df["Activity Level"]=np.select(Activecondition,Activity)
df
```

```python
df.to_csv('New_Covid19.csv')
```
